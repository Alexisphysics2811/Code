```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: 125*(-np.sin(((-4*np.pi)*(x**3))))+125
g = lambda x: 125*(np.sin((((-4*np.pi)*(x**2))/(x-3.25))))+125
```

Graph 1 - Version 1

```python
x = np.linspace(0,3,1000)
y = f(x)
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y)
plt.title('Gráfica de la función horizontal')
plt.xlabel("Grados")
plt.ylabel("Amplitud")
```
